```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np 
```

### Loading everything
- [x] trained model
- [x] scaler pickle
- [x] one hot encoder pickle

```python
model = load_model('model.h5') #loading the model
```

```python
with open('Label_encoder_gender.pkl', 'rb') as file:  #loading Geoencoder
    Label_encoder_gender = pickle.load(file)

with open('ohe_geo.pkl', 'rb') as file:  #loading Geography one hot encoder
    ohe_geo = pickle.load(file)

with open('scaler.pkl', 'rb') as file: #loding scaler
    scaler  = pickle.load(file)
```

```python
input_data = {

'CreditScore':600,
'Geography':'France',
'Gender': 'Male',
'Age': 40,
'Tenure': 3,
'Balance': 60000,
'NumOfProducts': 2,
'HasCrCard': 1,
'IsActiveMember': 1,
'EstimatedSalary': 50000

}
```

```python
geo_encoded = ohe_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns = ohe_geo.get_feature_names_out())
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender'] = Label_encoder_gender.transform(input_df['Gender'])
```

```python
input_df = pd.concat([input_df.drop('Geography', axis=1), geo_encoded_df], axis=1)
```

```python
input_df
```

```python
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_prob = prediction[0][0]
if prediction_prob > 0.5:
    print('Customer is likely to churn')
else:
    print('Customer is not likely to churn')
```
